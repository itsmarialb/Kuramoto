```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)


# explicitly set positions
pos = {1:(0,1), 2:(-1,0), 3:(1,0)}

options = {
    "font_size": 26,
    "node_size": 1500,
    "edgecolors": "black",
    "linewidths": 4,
    "width": 4,
}

val_map = {1: 0.0,
           2: 7.0,
           3: 10.0}

values = [val_map.get(node, 1.0) for node in G.nodes()]
nx.draw_networkx(G, pos, cmap=plt.get_cmap('Spectral'), node_color=values, with_labels=True, font_color='white',**options)


# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
#plt.show()
plt.savefig('graph.png')
```

```python
A = nx.adjacency_matrix(G)
print(A)
D = np.diag(np.ones(3)*2)
print(D)
L = D - A
print(L)
```

```python
from numpy.linalg import eig
w,v=eig(L)
print('E-value:', w)
print('E-vector', v)
```

```python
N = 3 # number of oscillators
steps = 100000

time = np.zeros(steps)
theta = np.zeros((steps,N))
order_p = np.zeros(steps)

lines = 0 # counter
with open("results_numerical.txt","r") as file1:
    for line in file1:
        line = line.split(' ')
        time[lines] = float(line[0])
        theta[lines]=line[1:4:]
        order_p[lines]=line[4]
        lines+=1 
```

```python
print(theta[0])
print(theta[5000])
print(theta[int(steps/2.)])
print(theta[steps-1])
```

```python
plt.figure(figsize=(5,5), dpi=80)

r=1
phi = np.linspace(-np.pi,np.pi,1000)
xc = r*np.cos(phi)
yc = r*np.sin(phi)
plt.plot(xc,yc,'--')

G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)


# explicitly set positions
pos = {1: (np.cos(theta[0][0]), np.sin(theta[0][0])), 2: (np.cos(theta[0][1]), np.sin(theta[0][1])), 3: (np.cos(theta[0][2]), np.sin(theta[0][2]))}

options = {
    "font_size": 26,
    "node_size": 1500,
    #"node_color": "white",
    "edgecolors": "black",
    "linewidths": 4,
    "width": 4,
}
val_map = {1: 0.0,
           2: 7.0,
           3: 10.0}

values = [val_map.get(node, 1.0) for node in G.nodes()]
nx.draw_networkx(G, pos, cmap=plt.get_cmap('Spectral'), node_color=values, with_labels=True, font_color='white',**options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)

plt.axis("off")
plt.ylim((-1.2,1.2))
plt.xlim((-1.2,1.2))
plt.savefig('estat_inicial.png')
plt.show()
```
